```python
import pandas as pd

# Load dataset
file_path = 'C:/Users/jonny/combined_tornado_enso_data.csv'
data = pd.read_csv(file_path)

# Tornado freqency by ENSO phase
tornado_frequency_by_enso = data['ENSO_Condition'].value_counts()

# Other tornado characteristics on ENSO phase
average_intensity_by_enso = data.groupby('ENSO_Condition')['mag'].mean()
average_size_by_enso = data.groupby('ENSO_Condition')['wid'].mean()
average_path_length_by_enso = data.groupby('ENSO_Condition')['len'].mean()

# Nocturnal tornado frequency by ENSO phase
nocturnal_tornadoes_by_enso = data.groupby('ENSO_Condition')['Is_Nocturnal'].mean()

# States with tornado frequency discrepancy during El Nino/La Nina
states_tornadoes_el_nino = data[data['ENSO_Condition'] == 'El Nino']['st'].value_counts()
states_tornadoes_la_nina = data[data['ENSO_Condition'] == 'La Nina']['st'].value_counts()

# Displaying states with more tornadoes during El Nino/La Nina
top_states_el_nino = states_tornadoes_el_nino.head(5)
top_states_la_nina = states_tornadoes_la_nina.head(5)

# Print the results
print("Tornado Frequency by ENSO Phase:", tornado_frequency_by_enso, sep='\n')
print("\nAverage Tornado Intensity by ENSO Phase:", average_intensity_by_enso, sep='\n')
print("\nAverage Tornado Size (Width) by ENSO Phase:", average_size_by_enso, sep='\n')
print("\nAverage Tornado Path Length by ENSO Phase:", average_path_length_by_enso, sep='\n')
print("\nFrequency of Nocturnal Tornadoes by ENSO Phase:", nocturnal_tornadoes_by_enso, sep='\n')
print("\nTop 5 States with Most Tornadoes During El Ni単o:", top_states_el_nino, sep='\n')
print("\nTop 5 States with Most Tornadoes During La Ni単a:", top_states_la_nina, sep='\n')
```
